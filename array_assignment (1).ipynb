# 1. Given an array, check if it contains any duplicates or not.
arr = [1, 2, 4, 2, 5, 9]
Output = True

```python
def contains_duplicates(arr):
    frequency = {}
    for num in arr:
        if num in frequency:
            return True
        frequency[num] = 1
    return False

# Example usage:
arr = [1, 2, 4, 2, 5, 9]
print(contains_duplicates(arr))  # Output: True
```

# 2. Given an array and an integer k, rotate the array to the right by k steps.
arr = [1, 2, 3, 4, 5, 6, 7] k = 3
Output = [5, 6, 7, 1, 2, 3, 4]

```python
def rotate_array(arr, k):
    # Calculate the effective rotation steps (in case k is larger than the array length)
    k = k % len(arr)
    
    # Rotate the array using slicing
    arr[:] = arr[-k:] + arr[:-k]

# Example usage:
arr = [1, 2, 3, 4, 5, 6, 7]
k = 3
rotate_array(arr, k)
print(arr)  # Output: [5, 6, 7, 1, 2, 3, 4]
```

# 3. Reverse the given array in-place, means without using any extra data structure.
arr = [2, 4, 5, 7, 9, 12]
Output = [12, 9, 7, 5, 4, 2]

```python
def reverse_array(arr):
    # Use Python slicing to reverse the array in-place
    arr[:] = arr[::-1]

# Example usage:
arr = [2, 4, 5, 7, 9, 12]
reverse_array(arr)
print(arr)  # Output: [12, 9, 7, 5, 4, 2]
```

# 4. Given an array of integers, find the maximum element in an array
arr = [10, 5, 20, 8, 15]
Output = 20

```python
def find_max(arr):
    if not arr:
        return None  # Handle empty array case
    max_element = arr[0]
    for num in arr:
        if num > max_element:
            max_element = num
    return max_element

# Example usage:
arr = [10, 5, 20, 8, 15]
max_element = find_max(arr)
print(max_element)  # Output: 20
```

# 5. Given a sorted array, remove the duplicate element without using any extra data structure.
arr = [1, 1, 2, 2, 2, 3, 3, 4, 4, 4, 5, 5]
Output = [1, 2, 3, 4, 5]

```python
def remove_duplicates(arr):
    if not arr:
        return []
    
    unique_index = 0
    for i in range(1, len(arr)):
        if arr[i] != arr[unique_index]:
            unique_index += 1
            arr[unique_index] = arr[i]
    
    # Truncate the array to keep only unique elements
    del arr[unique_index + 1:]

# Example usage:
arr = [1, 1, 2, 2, 2, 3, 3, 4, 4, 4, 5, 5]
remove_duplicates(arr)
print(arr)  # Output: [1, 2, 3, 4, 5]
```
